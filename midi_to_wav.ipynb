```python
import subprocess
import os
import time

# Ğ˜ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞµÑ‚ÑÑ GeneralUser-GS.sf2 Ğ² ĞºĞ°Ñ‡ĞµÑÑ‚Ğ²Ğµ Ğ·Ğ²ÑƒĞºĞ¾Ğ²Ğ¾Ğ³Ğ¾ ÑˆÑ€Ğ¸Ñ„Ñ‚Ğ°
SOUNDFONT_PATH = os.path.abspath("GeneralUser-GS.sf2")
SAMPLE_RATE = 44100
MIDI_FOLDER = "lettersabc/a/balanced_instruments/no_pauses/valid"
OUTPUT_DIR = os.path.join(MIDI_FOLDER, "wav")
os.makedirs(OUTPUT_DIR, exist_ok=True)

print(f"ğŸ“ Ğ¢ĞµĞºÑƒÑ‰Ğ°Ñ Ñ€Ğ°Ğ±Ğ¾Ñ‡Ğ°Ñ Ğ´Ğ¸Ñ€ĞµĞºÑ‚Ğ¾Ñ€Ğ¸Ñ: {os.getcwd()}")
print(f"ğŸµ SoundFont Ğ¿ÑƒÑ‚ÑŒ: {SOUNDFONT_PATH}")

for file in os.listdir(MIDI_FOLDER):
    if not file.lower().endswith(".mid"):
        continue

    midi_path = os.path.abspath(os.path.join(MIDI_FOLDER, file))
    output_wav = os.path.abspath(os.path.join(OUTPUT_DIR, os.path.splitext(file)[0] + ".wav"))

    print(f"ğŸ¼ ĞĞ°Ñ‡Ğ°Ñ‚Ğ° ĞºĞ¾Ğ½Ğ²ĞµÑ€Ñ‚Ğ°Ñ†Ğ¸Ñ: {midi_path}")
    cmd = f'fluidsynth -ni -F "{output_wav}" -r {SAMPLE_RATE} "{SOUNDFONT_PATH}" "{midi_path}"'

    try:
        start = time.time()
        with subprocess.Popen(cmd, shell=True, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True) as proc:
            for line in proc.stdout:
                print("ğŸ§¾", line.strip())
            proc.wait(timeout=60)
        duration = time.time() - start
        print(f"âœ… Ğ£ÑĞ¿ĞµÑˆĞ½Ğ¾: {output_wav} (â±ï¸ {duration:.2f} ÑĞµĞº)")
    except subprocess.TimeoutExpired:
        proc.kill()
        print(f"â° Ğ¢Ğ°Ğ¹Ğ¼Ğ°ÑƒÑ‚: {file}")
    except Exception as e:
        print(f"âŒ ĞÑˆĞ¸Ğ±ĞºĞ°: {file}: {e}")
```
